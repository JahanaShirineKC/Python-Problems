```python
#Count Frequency of Characters

def count_freq(n):
    res = {}
    for i in n:
        if i not in res:
            res[i] = n.count(i)    
    return res
    
n= 'banana'
print(count_freq(n))
```

```python
#Merge Two Dictionaries - without modifying orginal

d1 = {'a':1, 'b':2}
d2 = {'c':3, 'd':4}

res = d1 | (d2)
print(res)
```
