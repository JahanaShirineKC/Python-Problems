```python
a = 1
try:
    b = int(input("Enter a number"))
    result = a/b
    print(result)
except ZeroDivisionError:
    print("b cant be 0") # if b is '0' 
except Exception as e:
    print(f"Error : {e}") #  b is an alphabet
```

```python
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("Invalid input")
else:
    print("You entered:", num) #ValueError : when an inappropriate value is used within the code.
```
