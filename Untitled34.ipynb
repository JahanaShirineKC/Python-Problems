```python
import matplotlib.pyplot as plt

def plot_points():
    x_points = []
    y_points = []

    print("Enter points (x, y). Type 'done' to finish.")

    while True:
        user_input = input("Enter x and y coordinates separated by a space: ")
        if user_input.lower() == 'done':
            break
        try:
            x, y = map(float, user_input.split())
            x_points.append(x)
            y_points.append(y)
        except ValueError:
            print("Invalid input. Please enter two numeric values separated by a space.")

    plt.scatter(x_points, y_points, color='blue', label='User Points')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("User Input Points Plot")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    plot_points()
```
